```python
import matplotlib.pyplot as plt  # Importing the matplotlib library for plotting graphs

def plot_line_and_check_point(m, c, point):
    # Function to plot a line (y = mx + c) and check if a given point is above, below, or on the line
    
    # Line equation: y = mx + c
    x_values = range(-10, 11)  # Generate x-values for the line from -10 to 10
    y_values = [m * x + c for x in x_values]  # Compute corresponding y-values using the line equation

    # Plotting the line
    plt.plot(x_values, y_values, label=f'y = {m}x + {c}')  # Plot the line and add a label for it

    # Plotting the point
    x_point, y_point = point  # Unpack the coordinates of the given point
    plt.scatter(x_point, y_point, color='red', label=f'Point ({x_point}, {y_point})')  # Plot the point in red

    # Determine if the point is above, below, or on the line
    y_on_line = m * x_point + c  # Calculate the y-coordinate on the line at x_point
    if y_point > y_on_line:  # Check if the point is above the line
        position = "above"
    elif y_point < y_on_line:  # Check if the point is below the line
        position = "below"
    else:  # Check if the point lies exactly on the line
        position = "on"

    # Displaying the result on the plot
    plt.text(x_point, y_point, f'({x_point}, {y_point}) is {position} the line', fontsize=12, ha='left')
    # Annotate the graph with the position of the point relative to the line

    # Labels and title
    plt.xlabel('x')  # Label for the x-axis
    plt.ylabel('y')  # Label for the y-axis
    plt.axhline(0, color='black', linewidth=0.5)  # Add a horizontal line at y = 0 (x-axis)
    plt.axvline(0, color='black', linewidth=0.5)  # Add a vertical line at x = 0 (y-axis)
    plt.grid(True)  # Enable grid lines for better visualization
    plt.legend()  # Display the legend to identify the line and point
    plt.title(f'Point relative to the line y = {m}x + {c}')  # Add a title to the plot

    # Show plot
    plt.show()  # Display the plot

    return position  # Return whether the point is above, below, or on the line

# Example usage
m = 2  # Slope of the line
c = 5  # Y-intercept of the line
point = (5, 10)  # The point to check (x, y)

position = plot_line_and_check_point(m, c, point)  # Call the function and store the result
print(f"The point {point} is {position} the line y = {m}x + {c}")  # Print the result
```
